```python
def extract_ne(quote):
  words = word_tokenize(quote)
  tags = nltk.pos_tag(words)
  tree = nltk.ne_chunk(tags, binary=True)
  return set(" ".join(i[0] for i in t)
            for t in tree
            if hasattr(t, "label") and t.label() == "NE"
            )

```

## Comparison of Outputs and Discussion on Lemmatization in Healthcare NLP

### Tokenization Comparison:
- **NLTK Word Tokenization:** Separates words and punctuation. For example, `65-year-old` is treated as a single token, and `.` is a separate token.
- **spaCy Word Tokenization:** Offers more granular tokenization, splitting hyphenated words (e.g., `65-year-old` becomes `65`, `-`, `year`, `-`, `old`) and treating punctuation separately. This can be beneficial for deeper linguistic analysis, especially in medical texts where compound terms are common.
- **NLTK Sentence Tokenization:** Accurately identifies sentence boundaries.
- **spaCy Sentence Tokenization:** Also accurately identifies sentence boundaries, often performing well with complex sentence structures.

### Stemming vs. Lemmatization Comparison:

**Stemming (NLTK: Porter, Snowball, Lancaster):**
- **Porter Stemmer:** Aggressive, often reducing words to a common root that might not be a dictionary word (e.g., `radiating` -> `radiat`, `facilities` -> `facil`).
- **Snowball Stemmer:** Slightly less aggressive than Porter, but still produces non-dictionary words (e.g., `radiating` -> `radiat`, `facilities` -> `facil`).
- **Lancaster Stemmer:** The most aggressive, leading to highly shortened and often unrecognizable stems (e.g., `radiating` -> `rady`, `facilities` -> `facil`).

**Lemmatization (NLTK: WordNetLemmatizer, spaCy):**
- **NLTK WordNetLemmatizer:** Attempts to reduce words to their base or dictionary form (lemma). However, without Part-of-Speech (POS) tags, it defaults to noun lemmatization, which can be less accurate. For example, `presented` remains `presented` (as it's a verb, but treated as a noun by default), `was` remains `wa`, and `tests` becomes `test`. `Facilities` becomes `facility`, and `events` becomes `event`.
- **spaCy Lemmatization:** Generally more robust as it inherently uses POS tagging during its NLP pipeline. This allows it to derive more accurate lemmas, including verbs. For instance, `presented` -> `present`, `was` -> `be`, `radiating` -> `radiate`, `tests` -> `test`, `facilities` -> `facility`, `events` -> `event`.

### Why Lemmatization is Critical in Healthcare NLP:

1.  **Precision and Accuracy:** Healthcare data often contains highly specific terminology, abbreviations, and clinical notes. Stemming, being a rule-based approach, can aggressively chop off word endings, leading to non-existent words or ambiguous roots. This can lose crucial medical context (e.g., `infection` and `infectious` might stem to the same root, but have different clinical implications).
    Lemmatization, by returning the dictionary form, ensures that words retain their meaning and can be accurately matched to medical ontologies, terminologies (like SNOMED CT, ICD-10), and knowledge graphs.

2.  **Information Retrieval:** When searching electronic health records (EHRs) or medical literature, precise term matching is vital. If a search query uses a different inflection of a word than what's stored, stemming might retrieve irrelevant results or miss relevant ones. Lemmatization helps normalize variations of medical terms (e.g., `diagnoses`, `diagnosing`, `diagnosed` all become `diagnosis`), leading to more comprehensive and accurate information retrieval.

3.  **Clinical Decision Support:** In systems designed to assist clinicians, misinterpreting a word due to aggressive stemming could lead to incorrect recommendations or missed alerts. For example, `medication` vs. `medicated` have distinct meanings. Lemmatization helps maintain the semantic integrity required for reliable clinical applications.

4.  **Feature Engineering for Machine Learning:** For tasks like disease prediction, risk assessment, or identifying adverse drug events, machine learning models rely on high-quality features. Lemmatized words provide a cleaner, more semantically consistent input for these models, improving their performance and interpretability. This reduces the feature space while retaining meaning, which is critical for complex medical vocabulary.

5.  **Interoperability:** In a field where data sharing and standardization are paramount, lemmatization promotes consistency across different systems and datasets by reducing word forms to a common base. This aids in data integration and analysis across various healthcare platforms.

In summary, while stemming is faster, its lack of linguistic intelligence makes it less suitable for the nuanced and critical domain of healthcare. Lemmatization, especially with POS tagging capabilities like spaCy, provides a more accurate and context-aware normalization, which is indispensable for building robust and reliable NLP applications in healthcare.
