```python
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
text = """Loving the new AI features


 #AI #MachineLearning"""
doc = nlp(text)
nouns = []
verbs = []
for token in doc:
    if token.pos_ in ["NOUN", "PROPN"]:
        nouns.append(token.text)
    elif token.pos_ == "VERB":
        verbs.append(token.text)
noun_freq = Counter(nouns)
verb_freq = Counter(verbs)
```

```python
print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
```

```python
essay_text2 = """Artificial Intelligence is an emerging field of computer science that focuses on creating intelligent machines.
It enables systems to learn from data, make decisions, and solve complex problems efficiently.
AI applications are widely used in education, healthcare, transportation, and research sectors."""
```

```python
tokens=word_tokenize(essay_text2)
print(tokens)
```

```python
nltk_pos_tags=nltk.pos_tag(tokens)
print(nltk_pos_tags)
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(essay_text2)

spacy_pos_tags = [(token.text, token.pos_) for token in doc]
print(spacy_pos_tags)
```

```python
nltk_nouns = [word for word, tag in nltk_pos_tags if tag.startswith('NN')]
print(nltk_nouns)
```

```python
spacy_nouns = [token.text for token in doc if token.pos_ in ["NOUN", "PROPN"]]
print(spacy_nouns)
```

```python
nltk_verbs = [word for word, tag in nltk_pos_tags if tag.startswith('VB')]
print(nltk_verbs)
```

```python
spacy_verbs = [token.text for token in doc if token.pos_ in ["VERB", "AUX"]]
print(spacy_verbs)
```

```python
noun_frequency = Counter(spacy_nouns)
verb_frequency = Counter(spacy_verbs)

print("Noun Frequency:", noun_frequency)
print("Verb Frequency:", verb_frequency)
```
