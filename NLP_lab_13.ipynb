Text Classification I: Implementing Baselines (Naive Bayes with Scikit-learn)

1. Environment Setup


Import Libraries

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```


Dataset

```python
import pandas as pd

data = {
    'text': [
        "Congratulations! You've won a free ticket",
        "Claim your reward now",
        "Exclusive deal just for you",
        "Hi, are we still on for dinner?",
        "Don't forget the meeting at 10 AM",
        "Can you send me the report?",
        "Get rich quick scheme",
        "Urgent! Update your bank details",
        "See you at the gym",
        "Happy birthday! Have a great day"
    ],
    'label': [
        1, 1, 1,   # Spam
        0, 0, 0,   # Not Spam
        1, 1, 0, 0
    ]
}

df = pd.DataFrame(data)
print(df)
```


Feature Extraction (Bag-of-Words)

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text'])

y = df['label']
```


Data Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```


Train Naive Bayes Model


```python
model = MultinomialNB()
model.fit(X_train, y_train)
```


Model Testing

```python
y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Actual:", y_test.values)

```

Model Evaluation

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```


Model Saving

```python
import joblib

# Save trained model
joblib.dump(model, 'naive_bayes_model.pkl')

# Save vectorizer also (VERY IMPORTANT)
joblib.dump(vectorizer, 'vectorizer.pkl')

```


Load Model

```python

# Load model
loaded_model = joblib.load('naive_bayes_model.pkl')

# Load vectorizer
loaded_vectorizer = joblib.load('vectorizer.pkl')
```


Model Deployment

```python
text = ["Free offer now"]

# Transform using saved vectorizer
text_vec = loaded_vectorizer.transform(text)

# Predict
prediction = loaded_model.predict(text_vec)

print("Prediction:", prediction[0])
```

Case Study - 2 : News Group Dataset

Import Libraries

```python

import numpy as np
import pandas as pd

# Sklearn modules
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# Evaluation metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

Load Dataset

```python
 #Load dataset (subset for binary classification)
categories = ['sci.space', 'rec.sport.hockey']
data = fetch_20newsgroups(subset='all', categories=categories)

texts = data.data
labels = data.target

# Display sample
for i in range(3):
    print("Text:", texts[i][:200])
    print("Label:", labels[i])
    print("------")
```


Text Preprocessing

```python
import re

def preprocess(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # remove punctuation
    return text

texts = [preprocess(t) for t in texts]

```


Feature Extraction

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)
```

Data Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)
```

Model Training

```python
model = MultinomialNB()
model.fit(X_train, y_train)

```


Model Testing

```python
y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Actual:", y_test)
```


Model Evaluation

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

Model Saving

```python
import joblib

# Save trained model
joblib.dump(model, 'naive_bayes_model_news.pkl')

# Save vectorizer also (VERY IMPORTANT)
joblib.dump(vectorizer, 'vectorizer_news.pkl')
```

Load Model

```python
# Load model
loaded_model = joblib.load('naive_bayes_model_news.pkl')

# Load vectorizer
loaded_vectorizer = joblib.load('vectorizer_news.pkl')
```

Model Deployment

```python
test_samples = [
    "NASA launched a new satellite into space",
    "The hockey team won the championship game",
    "3D rendering and computer graphics techniques improved"
]

# Transform
test_vec = loaded_vectorizer.transform(test_samples)

# Predict
predictions = loaded_model.predict(test_vec)

# Map labels to category names
for text, pred in zip(test_samples, predictions):
    print("\nText:", text)
    print("Predicted Category:", data.target_names[pred])
```
